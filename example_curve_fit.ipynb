```python
import numpy as np
import scipy as sc
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
```

```python
y = np.array([x*x for x in range(1, 20)])
x = np.array([i for i in range(1, 20)])

def func_linear(x, a, b, c):
  return a*x + b
def func_quadro(x, a, b, c):
  return a*x*x + b*x + c
def func_exp(x, a, b, c):
  return a*(np.exp(x)) + b


params, cov = curve_fit(func_linear, x, y)

fig, ax1 = plt.subplots(figsize=(5, 3), dpi=100)
plt.plot(x, y, 'o', color='k', markersize=3)
plt.plot(x, func_linear(x, params[0], params[1], params[2]), 'r', linestyle='-')
```
